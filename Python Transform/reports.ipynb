```python
df = pd.read_csv('./raw/газпром_АО_марк.csv', sep=';', header=0)
```

```python
claim_per_agent = df['Оператор'].value_counts().sort_index()
claim_per_agent.to_clipboard()
```

```python
# Assuming your DataFrame is named 'df'
# Create a list of marker numbers to check for
marker_numbers = [f" {i}\." for i in range(1, 16)]

# Create new columns for each topic and fill with boolean values
for i, marker_number in enumerate(marker_numbers, start=1):
    column_name = f"Topic {i}"
    df[column_name] = df['Маркер'].str.contains(marker_number)
```

```python
pivot_table_topics = df.pivot_table(index='Регион', columns='Регион', values=['Topic 1', 'Topic 2'], aggfunc='sum')
```

```python
pivot_table_topics
```

```python
topic_columns = ['Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'Topic 5',
                 'Topic 6', 'Topic 7', 'Topic 8', 'Topic 9', 'Topic 10',
                 'Topic 11', 'Topic 12', 'Topic 13', 'Topic 14', 'Topic 15']

sum_by_region = df.groupby('Регион')[topic_columns].sum()
```

```python
sum_by_region.to_clipboard()
```
